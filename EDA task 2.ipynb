```python
# checking wether the values changed or not
cricket.head()
```

```python
# filling the nan value 
cricket.fillna(0, inplace = True)
#checking that nan values are filled or not
cricket.head()
```

```python
# 1. Create 'Not_Out' column
# here'regex' tells the computer to treat the star as a special pattern 
# na=False: If a score is missing (like an empty cell), it will be marked as "not containing a star" (which is correct).
cricket['Not_Out'] = cricket['Highest_Score'].str.contains('\*', regex=True, na=False)

# 2. Remove '*' from 'Highest_Score'
cricket['Highest_Score'] = cricket['Highest_Score'].str.replace('\*', '', regex=True)

# 3. Convert 'Highest_Score' to Numeric
#errors='coerce': If it finds something that's not a number (after removing the star), 
#it will replace it with a special "not a number" value (called NaN)
#This helps avoid errors later.
cricket['Highest_Score'] = pd.to_numeric(cricket['Highest_Score'], errors='coerce')

# Handle NaN values
cricket['Highest_Score'].fillna(0, inplace=True)
```

```python
# Verify the changes
cricket.head()
```

```python
# cricket data that tells you the best bowling performance in a match, like "4/25" (meaning 4 wickets taken for 25 runs). 
#This code helps you separate those two numbers (wickets and runs) into two new, separate columns.

def extract_bowling_figures(value):
    """Extracts wickets and runs from the 'Best_Bowling_Match' string."""
    try:
        if isinstance(value, str):
            parts = value.split('/')
            if len(parts) == 2:
                wickets = int(parts[0])
                runs = int(parts[1])
                return wickets, runs
    except (ValueError, AttributeError):
        pass  # Handle errors if conversion fails or value is not a string
    return np.nan, np.nan  # Return NaN if extraction fails

# Apply the function to create new columns
cricket[['Best_Bowling_Wickets', 'Best_Bowling_Runs']] = cricket['Best_Bowling_Match'].apply(extract_bowling_figures).apply(pd.Series)

# Handle NaN values (fill with 0 for now)
cricket[['Best_Bowling_Wickets', 'Best_Bowling_Runs']] = cricket[['Best_Bowling_Wickets', 'Best_Bowling_Runs']].fillna(0)
cricket = cricket.drop('Best_Bowling_Match', axis=1) 
```

```python
# 1. BATTING PERFORMANCE
#How does the distribution of 'Runs_Scored' vary across different 'Years'? 

#Boxplots to Visualize Distribution:
#Boxplots are great for showing the distribution of a numerical variable across different categories (in this case, years). 
#They'll help us see how the median, quartiles, and outliers of 'Runs_Scored' change over the years.

# Create the box plot using Matplotlib
fig, ax = plt.subplots(figsize=(10, 6))  # Adjust figure size as needed
# Extract the years and runs scored
years = cricket['Year'].unique()
data = [cricket['Runs_Scored'][cricket['Year'] == year] for year in years]
# Create the boxplot
ax.boxplot(data, labels=years)
# Set the title and labels
ax.set_title('Distribution of Runs Scored Across Years')
ax.set_xlabel('Year')
ax.set_ylabel('Runs Scored')

# Display the plot
plt.show()
```

```python
#to find the yearly avg of runs scored  we need to change data type of runs_scored
cricket['Runs_Scored'] = pd.to_numeric(cricket['Runs_Scored'], errors='coerce').astype(float)
print(cricket['Runs_Scored'].head(10))
```

```python
#Are there any noticeable trends in scoring patterns over time?

#"Noticeable trends" means identifying any consistent or significant changes in the scoring patterns from year to year. 
#Are scores generally increasing? Decreasing? Are certain years unusually high or low scoring?
#"Scoring patterns" refers to the overall characteristics of how runs are scored. 

#Line Plot to Visualize Yearly Average:
#A line plot will show the trend of the average 'Runs_Scored' over time.

# Calculate the yearly average of 'Runs_Scored' using pandas
yearly_avg_runs = cricket.groupby('Year')['Runs_Scored'].mean().reset_index()
# Create the line plot using Matplotlib
fig, ax = plt.subplots(figsize=(10, 6))  # Adjust figure size as needed
# Plot the 'Year' on the x-axis and 'Runs_Scored' on the y-axis
ax.plot(yearly_avg_runs['Year'], yearly_avg_runs['Runs_Scored'], marker='o', linestyle='-') # Added marker and line style
# Set the title and labels
ax.set_title('Yearly Average Runs Scored')
ax.set_xlabel('Year')
ax.set_ylabel('Average Runs Scored')
# Add gridlines for better readability
ax.grid(True)
# Display the plot
plt.show()
```

```python
#2.BOWLING PERFORMANCE
#"How does the 'Economy_Rate' vary across different 'Years'?"

#'Economy_Rate':In cricket, the economy rate is a measure of how many runs a bowler concedes for every over they bowl.
# A lower economy rate is generally better for the bowling team because it means the bowler is giving away fewer runs.
#'across different 'Years': This means we need to look at the economy rates for each year in your IPL dataset separately.

# Create the box plot using Matplotlib
fig, ax = plt.subplots(figsize=(10, 6))  # Adjust figure size as needed

# Extract the years and economy rates
years = cricket['Year'].unique()
data = [cricket['Economy_Rate'][cricket['Year'] == year] for year in years]

# Create the boxplot
ax.boxplot(data, labels=years)

# Set the title and labels
ax.set_title('Distribution of Bowling Economy Rate Across Years')
ax.set_xlabel('Year')
ax.set_ylabel('Economy Rate')
# Display the plot
plt.show()
```

```python
#"Has bowling become more or less economical over time?"
#Over all the IPL seasons, have bowlers generally gotten better at keeping the run rate down (lower economy rate), 
#or have they gotten worse (higher economy rate)?

# Calculate the yearly average Economy Rate using pandas
yearly_avg_economy = cricket.groupby('Year')['Economy_Rate'].mean().reset_index()

# Create a simple line plot to show the trend over time using Matplotlib
fig, ax = plt.subplots(figsize=(10, 6))  # Adjust figure size as needed

# Plot the 'Year' on the x-axis and 'Economy_Rate' on the y-axis
ax.plot(yearly_avg_economy['Year'], yearly_avg_economy['Economy_Rate'], marker='o', linestyle='-') # Added marker and line style
# Set the title and labels
ax.set_title('Trend of Average Bowling Economy Rate Over Time')
ax.set_xlabel('Year')
ax.set_ylabel('Average Economy Rate')
# Add gridlines for better readability
ax.grid(True)
# Display the plot
plt.show()
```

```python
# 3. All Rounder performance
#"How consistent are players' performances over different 'Years'?
#Can we identify players with stable batting or bowling averages?

#each cricket player performs in the IPL across all the seasons they've played. 
#"Consistent performance" means a player performs at a similar level from one year to the next.

# For Batting Consistency (using Batting Average as an example):
batting_consistency = cricket.groupby(['Player_Name', 'Year'])['Batting_Average'].mean().reset_index()

# Create the line plot using Matplotlib
fig, ax = plt.subplots(figsize=(12, 6))  # Adjust figure size as needed

# Get unique player names
players = batting_consistency['Player_Name'].unique()
# Plotting each player's batting average
for player in players:
    player_data = batting_consistency[batting_consistency['Player_Name'] == player]
    ax.plot(player_data['Year'], player_data['Batting_Average'], marker='o', linestyle='-', label=player)
# Set the title and labels
ax.set_title('Batting Average Consistency Over Years')
ax.set_xlabel('Year')
ax.set_ylabel('Batting Average')
ax.legend(title='Player Name')
ax.grid(True)  # Add gridlines

# Show the plot
plt.show()
```

```python
# For Bowling Consistency (using Bowling Average as an example):
bowling_consistency = cricket.groupby(['Player_Name', 'Year'])['Bowling_Average'].mean().reset_index()

# Create the line plot using Matplotlib
fig, ax = plt.subplots(figsize=(12, 6))  # Adjust figure size as needed

# Get unique player names
players = bowling_consistency['Player_Name'].unique()

# Plotting each player's bowling average
for player in players:
    player_data = bowling_consistency[bowling_consistency['Player_Name'] == player]
    ax.plot(player_data['Year'], player_data['Bowling_Average'], marker='o', linestyle='-', label=player)

# Set the title and labels
ax.set_title('Bowling Average Consistency Over Years')
ax.set_xlabel('Year')
ax.set_ylabel('Bowling Average')
ax.legend(title='Player Name')
ax.grid(True)  # Add gridlines

# Show the plot
plt.show()
```
